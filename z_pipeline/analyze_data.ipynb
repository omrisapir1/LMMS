```python
def get_questions(greedy_tokens):
    try:
        start_index = greedy_tokens.rindex('<|endoftext|>') + 13
        end_index = greedy_tokens.index("assistant\n") + 10
        return greedy_tokens[start_index: end_index]
    except:
        return None
```

```python
df['question'] = df['greedy_tokens'].apply(get_questions)
```

```python
df = df[~df['question'].isnull()]
```

```python
answer_token_id = 151665
pad_id = 151643
def get_answer(text):
    input_ids = torch.Tensor([tokenizer.encode(text)]).to(torch.long).to('cuda')
    full_attn = (input_ids != pad_id).long()
    out_full = model(input_ids = input_ids)
    hidden_last = out_full.hidden_states[-1]  # [B, T_total, H]

    mask = (input_ids == answer_token_id)
    has_answer = mask.any(dim=1)

    # if some rows didn't generate <ANSWER>, you can decide a fallback:
    # e.g., use last token position (or skip those rows)
    pos = mask.float().argmax(dim=1)
    pos = torch.where(
        has_answer,
        pos,
        (full_attn.sum(dim=1) - 1).clamp(min=0)  # fallback = last non-pad token
    )

    bidx = torch.arange(hidden_last.size(0), device=hidden_last.device)
    answer_hidden = hidden_last[bidx, pos]  # [B, H]

    digit_logits = torch.stack([head(answer_hidden) for head in model.digit_heads], dim=1)
    return digit_logits.argmax(dim=-1)
```

```python
import random
def generate_random_prompt(r):
    count_z = r['greedy_tokens'].count('Z_')
    return r['question'] + ''.join(f'<Z_{random.randint(0,1020)}>' for i in range(count_z)) + '<ANSWER>'
```

```python
df['rand_ans'] = df.apply(generate_random_prompt,axis=1)
```

```python
df['rand_ans_digit'] = df['rand_ans'].apply(get_answer)
```
